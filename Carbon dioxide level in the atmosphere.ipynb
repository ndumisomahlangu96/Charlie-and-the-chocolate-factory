```python
year = [
        1980, 1982, 1984, 1986, 1988, 
        1990, 1992, 1994, 1996, 1998, 
        2000, 2002, 2004, 2006, 2008, 
        2010, 2012
]

CO2_level = [
    338.7, 341.2, 344.4, 347.2, 351.5,
    354.2, 356.3, 358.6, 362.4, 366.5,
    369.4, 373.2, 377.5, 381.9, 385.6,
    389.9, 393.8
                
]

print("year = ",year);
print("type = ", type(year));
print(" ");

print("CO2 levels = ", CO2_level);
print("type = ", type(CO2_level));
print(" ");

from matplotlib import pyplot as plt;
x = year;
y = CO2_level;
plt.scatter(x,y);
plt.xlabel("Years");
plt.ylabel("CO2 level(in ppm)");
plt.title("Maua Loa Observatory from 1980 to 2012");
plt.show();
```
